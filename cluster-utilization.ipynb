```python
import pandas as pd
import numpy as np
import glob
```

```python
experiment = "E2"
schedulers = ["default-scheduler"]
for scheduler in schedulers:
    rawdata_files = glob.glob("/home/goughes/k8s-scheduling/results/" + experiment + "/" + experiment + "-" + scheduler + "*-rawdata-sum.csv")
    dfs = (pd.read_csv(f) for f in rawdata_files)
    avg = pd.concat(dfs).groupby(level=0).mean()
    running = avg['all_running']
    running.plot()
    y = np.array(running)
    area = np.trapz(y, dx=1)
    total = 384 * len(running)
    print("area", area)
    print("total", total)
    print("percent", scheduler, area / total * 100)
```
