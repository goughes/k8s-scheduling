```python
import pandas as pd
import numpy as np
import glob
```

```python
cols = ["tenant3_running"] #"tenant2_running", "tenant3_running"]
file_paths = glob.glob(f"results//E1//E1-tenant3-default-scheduler-run*-rawdata.csv")
for file in file_paths:
    newname = file.split(".")[0] + "-sum.csv"
    print(newname)
    df = pd.read_csv(file)
    df['all_running'] = df[cols].sum(axis=1)
    df.to_csv(newname)
```

```python
experiment = "E1"
scheduler = "default-scheduler"
tenants = ['tenant1', 'tenant2', 'tenant3']
for tenant in tenants:
    file_paths = glob.glob(f"results//E1//E1-{tenant}-default-scheduler-run*-rawdata-sum.csv")
    dfs = (pd.read_csv(f) for f in file_paths)
    avg = pd.concat(dfs).groupby(level=0).mean()
    running = avg['all_running']
    running.plot()
    y = np.array(running)
    area = np.trapz(y, dx=1)
    total = 384 * len(running)
    print("area", area)
    print("total", total)
    print("percent", scheduler, tenant, area / total * 100)
```
